```python
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import string
```

```python
corpus = PlaintextCorpusReader('Files', '.*')
```

```python
all_text = corpus.raw()
print(len(all_text))
all_text[:100]
```

```python
words = corpus.words()
words
```

```python
len(words)
```

```python
stop_words = stopwords.words('english')
stop_words
```

```python
color_map = ListedColormap(['orange', 'green', 'red', 'magenta'])

cloud = WordCloud(background_color='white', colormap=color_map, stopwords=stop_words, max_words=100)
cloud.generate(all_text)

plt.imshow(cloud)
plt.show()
```

```python
words_without_stop_words = [word for word in words if word not in stop_words]
words_without_stop_words
```

```python
print(len(words))
print(len(words_without_stop_words))
```

```python
words_without_punct = [word for word in words_without_stop_words if word not in string.punctuation]
len(words_without_punct)
```

```python
frequency = nltk.FreqDist(words_without_punct)
frequency
```

```python
most_commons = frequency.most_common(100)
most_commons
```
