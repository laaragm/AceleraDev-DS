```python
from math import sqrt

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import scipy.stats as sct
import seaborn as sns
```

```python
#Quando H0 é realmente falsa, a distribuição do valor-p depende do poder do teste
#Quando H0 é realmente verdadeira, o valor-p tem distribuição uniforme
#H0 é falsa
p_values1 = []
for i in range(1000):
    p_values1.append(sct.ttest_1samp(sct.norm.rvs(loc=10,scale=5,size=100), popmean=9.5).pvalue)
    
p_values2 = []
for i in range(1000):
    p_values2.append(sct.ttest_1samp(sct.norm.rvs(loc=10, scale=5, size=100), popmean=9).pvalue)
    
fig, axs = plt.subplots(1,2,figsize=(20,10))
sns.distplot(p_values1, kde=False, bins=20, hist_kws={'density':True}, ax=axs[0])
sns.distplot(p_values2, kde=False, bins=20, hist_kws={"density": True}, ax=axs[1])
plt.show()
```

```python
#H0 é verdadeira
p_values = []
for i in range(1000):
    p_values.append(sct.ttest_1samp(sct.norm.rvs(loc=10, scale=5, size=100), popmean=10).pvalue)

sns.distplot(p_values, kde=False, bins=20, hist_kws={"density": True})
plt.show()
```

```python
#O q-q plot é um gráfico para comparação de distribuições de probabilidades.
#Q-Q plot para dados normais
normal_data = sct.norm.rvs(loc=10,scale=4,size=1000)

sm.qqplot(normal_data, fit=True, line='45')
plt.show()
```

```python
#Q-Q plot para dados não normais
non_normal_data = sct.expon.rvs(size=1000)

sm.qqplot(non_normal_data, fit=True, line="45")
plt.show()
```
