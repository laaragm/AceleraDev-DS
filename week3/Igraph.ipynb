```python
from igraph import Graph
from igraph import plot
```

```python
graph1 = Graph(edges = [(0,1), (1,2), (2,3), (3,0)], directed=True)
graph1.vs['label'] = range(graph1.vcount())
print(graph1)
plot(graph1, bbox=(200,200))
```

```python
graph2 = Graph(edges = [(0,1), (1,2), (2,3), (3,0), (0,3), (3,2), (2,1), (1,0)], directed=True)
graph2.vs['label'] = range(graph2.vcount())
plot(graph2, bbox=(200,200))
```

```python
graph3 = Graph(edges = [(0,1), (1,2), (2,3), (3,0), (1,1)], directed=True)
graph3.vs['label'] = range(graph3.vcount())
plot(graph3, bbox=(200,200))
```

```python
graph4 = Graph(edges = [(0,1), (1,2), (2,3), (3,0), (1,1)], directed=True)
plot(graph4, bbox=(200,200))
graph4.add_vertex(4)
graph4.vs['label'] = range(graph4.vcount())
graph4.add_edges([(4,1), (4,2)])
plot(graph4, bbox=(200,200))
```

```python
graph5 = Graph(directed=False)
graph5.add_vertices(10)
graph5.vs['label'] = range(graph5.vcount())
graph5.add_edges([(1,2), (1,3), (3,4), (5,6), (6,7), (4,3)])
plot(graph5, bbox=(200,200))
```

```python
graph6 = Graph(edges = [(0,1), (1,2), (2,3), (3,0), (1,1)], directed=False)
plot(graph6, bbox=(200,200))
graph6.add_vertex(4)
graph6.add_vertex(5)
graph6.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F']
graph6.add_edges([(4,1), (4,2)])
print(graph6.get_adjacency())
plot(graph6, bbox=(200,200))
```

```python
node_path = graph6.get_shortest_paths(0, 4, output='vpath')
edge_path = graph6.get_shortest_paths(0, 4, output='epath')
print(node_path)
print(edge_path)
```

```python
print(graph6.get_adjacency()[0])
print(graph6.get_adjacency()[0,1])
```

```python
for v in graph6.vs:
    print(v)
```

```python
graph7 = Graph(edges = [(0,1), (2,3), (0,2), (0,3)], directed=True)
graph7.vs['label'] = ['P1', 'P2', 'P3', 'P4']
graph7.vs['peso'] = [30,20,40,10]
plot(graph7, bbox=(200,200))
```

```python
graph7.es['TipoAmizade'] = ['Amigo', 'Amigo', 'No conhece', 'Amigo']
graph7.es['Valor'] = [1,2,0,1]
for e in graph7.es:
    print(e)
```

```python
plot(graph7, bbox=(200,200), vertex_size=graph7.vs['peso'], edge_width=graph7.es['Valor'])
```
