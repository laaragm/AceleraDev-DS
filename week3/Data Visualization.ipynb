```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import axes3d
```

```python
data = pd.read_csv('trees.csv')
data.head()
```

```python
#Histograma
plt.hist(data.iloc[:,1], bins=6)
plt.title('Trees')
plt.ylabel('Frequency')
plt.xlabel('Height')
plt.show()
```

```python
sns.distplot(data.iloc[:,1], hist=True, kde=True, bins=6, color='blue', hist_kws={'edgecolor': 'black'})
plt.show()
```

```python
base = pd.read_csv('co2.csv')
base.head()
```

```python
x = base.conc
y = base.uptake

unique = list(set(base.Treatment))
#print(unique)

for i in range(len(unique)):
    index = base.Treatment == unique[i]
    plt.scatter(x[index], y[index], label=unique[i])
    
plt.legend(loc='lower right')
plt.show()
```

```python
chicken = pd.read_csv('chicken.csv')
chicken.head()
```

```python
plt.figure()

plt.subplot(3,2,1)
sns.distplot(chicken.loc[chicken['feed'] == 'horsebean'].weight).set_title('horsebean')

plt.subplot(3,2,2)
sns.distplot(chicken.loc[chicken['feed'] == 'casein'].weight).set_title('casein')

plt.subplot(3,2,3)
sns.distplot(chicken.loc[chicken['feed'] == 'linseed'].weight).set_title('linseed')

plt.subplot(3,2,4)
sns.distplot(chicken.loc[chicken['feed'] == 'meatmeal'].weight).set_title('meatmeal')

plt.subplot(3,2,5)
sns.distplot(chicken.loc[chicken['feed'] == 'soybean'].weight).set_title('soybean')

plt.subplot(3,2,6)
sns.distplot(chicken.loc[chicken['feed'] == 'sunflower'].weight).set_title('sunflower')

plt.tight_layout()
plt.show()
```
